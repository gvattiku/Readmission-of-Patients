```python
claims_diag_proc_mem_dgn_seq = claims_diag_proc_mem.drop_duplicates(subset=['claim_id'], keep='first')
```

```python
claims_diag_proc_mem_dgn_seq = claims_diag_proc_mem_dgn_seq.sort_values(by=['time_days','patient_id','claim_id'])
```

```python
def comp_prev(a):
    if (str(a['patient_id'][1:])) == str(a['patient_id'][:-1]) and (int(a['claim_days'][1:])-int(a['claim_days'][:-1]) <300):
        return 1
    else:
        return 0
```

```python
claims_diag_proc_mem_dgn_seq['hosp_nxt_month'] = comp_prev(claims_diag_proc_mem_dgn_seq)
claims_diag_proc_mem_dgn_seq[claims_diag_proc_mem_dgn_seq['hosp_nxt_month'] == 1]
```

```python
start_day = 175
timeline = 6  # months
end_day = 175+6*30
next_30_days = end_day+30

claims_timeline = claims_diag_proc_mem_dgn_seq[claims_diag_proc_mem_dgn_seq["time_days"].between(start_day, end_day, inclusive = True)] 
claims_next_30 = claims_diag_proc_mem_dgn_seq[claims_diag_proc_mem_dgn_seq["time_days"].between(end_day, next_30_days, inclusive = True)]

claims_next_30['hosp_nxt_month'] = 1
#join two claim table and update the dependent variable
claims_next_30.to_sql("claims_next_30", con = conn, index = False)
claims_timeline.to_sql("claims_timeline", con = conn, index = False)
claims_timeline_merged = pd.read_sql("select claims_timeline.*,claims_next_30.hosp_nxt_month from claims_timeline "+
                                     "left join claims_next_30 on claims_next_30.claim_id = claims_timeline.claim_id ",conn)
```
